```python
# Using Selenium to extract google play store reviews

from selenium import webdriver
import time
from bs4 import BeautifulSoup

test_url = 'https://play.google.com/store/apps/details?id=com.news.hindutamil'

option = webdriver.ChromeOptions()
option.add_argument("--incognito")
chromedriver = r'chromedriver.exe'
browser = webdriver.Chrome(chromedriver, options=option)
browser.get(test_url)
html_source = browser.page_source
browser.close()
```

```python
# extracting authors

soup=BeautifulSoup(html_source, "html.parser")

review_author_list_src = soup.find_all('div', {'class','X5PpBb'})
review_author_name_list = []

for val in review_author_list_src:
    try:
        review_author_name_list.append(val.get_text())
    except:
        pass
review_author_name_list[:3]
```

```python
# extracting review dates

date_src = soup.find_all('span',{'class', 'bp9Aid'})

date_src
date_list = []

for val in date_src:
    date_list.append(val.get_text())
date_list[:3]
```

```python
# extracting review content

review_content_src = soup.find_all('div',{'class', 'h3YV2d'})
review_content_list = []

for val in review_content_src:
    review_content_list.append(val.get_text())
review_content_list[:3]
```

```python
# Rating

rating = soup.find_all('div',{'class', 'F7XJmb'})
ratings = []

for val in rating:
    ratings.append(val.get_text())
ratings[:3]
```

```python
data = pd.DataFrame({"review_content_list":review_content_list , 
                     "review_author_name_list":review_author_name_list, "date_list":date_list})
data
```
